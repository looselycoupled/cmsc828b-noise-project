````python
y = [22.85, 14.50, 10.29, 4.62, 3.87, 3.84]
huda = [27.2, 17.6, 11.2, 5.6, 3.2, 3.2]
labels = ['0%','5%','10%','20%','50%','100%']

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
barlist = ax.bar(x - (width/2),y, width=width, label="transformer")
barlist = ax.bar(x + (width/2),huda, color="lawngreen", width=width, label="rnn")

ax.set_xticks(x)
ax.set_xticklabels(labels)

ax.set_ylabel("BLEU score")
ax.set_xlabel("percentage of noise")
ax.set_title("BLEU vs Percentage of Noise")

plt.legend()
plt.show(0)
````

````python
total = 3004

total = 3004

copies = np.array([1, 12, 33, 129, 611, 1322, 1678, 2544, 1897]) / total 
copies = copies * 100
copies = copies.astype(int).tolist()

similar_copies = np.array([43, 69, 117, 244, 1024, 1777, 2720, 2940, 2952]) / total
similar_copies = similar_copies * 100
similar_copies = (similar_copies.astype(int) - copies).tolist()


labels = ['1%','2%','3%','4%','5%','10%','20%','50%','100%']

fig, ax = plt.subplots()
width = 0.35 

ax.bar(labels, copies, width, label='Copied Sentences')
ax.bar(labels, similar_copies, width, bottom=copies, label='Similar Sentences', color="lightsteelblue")

ax.set_ylabel('percentage of sentences')
ax.set_xlabel('amount of noise')
ax.set_title('Untranslated Target Copying')
ax.legend()

plt.show()

````
